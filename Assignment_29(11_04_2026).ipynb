#### Designed Solution: Helmet Detection System for Road Safety
- Problem Statement: Many accidents occur because riders do not wear helmets, leading to serious injuries and safety violations.
- Proposed Solution: Develop an object detection-based system that identifies whether a motorcyclist is wearing a helmet and flags violations automatically.

```python
from ultralytics import YOLO
import cv2

# Load trained model
model = YOLO("helmet_model.pt")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    # Draw results
    annotated_frame = results[0].plot()

    cv2.imshow("Helmet Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
