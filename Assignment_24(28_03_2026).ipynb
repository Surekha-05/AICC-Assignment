Try confusing an LLM with tricky prompts and document results.

```python
import pandas as pd
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

# Different tricky prompts
prompts = [
    "Is the sentence 'This statement is false' true or false?",
    "If you could see sound, what would it look like?",
    "Can an unstoppable force stop an immovable object?",
    "Write a sentence that has no meaning but looks meaningful.",
    "If you remove all numbers from time, how would you measure it?"
]

for i, prompt in enumerate(prompts):
    print(f"\nTest Case {i+1}")
    print("Prompt:", prompt)
    
    response = generator(prompt, max_length=60, num_return_sequences=1)
    
    print("Response:", response[0]['generated_text'])
    print("-" * 50)

results = []

for i, prompt in enumerate(prompts):
    response = generator(prompt, max_length=60)[0]['generated_text']
    
    results.append({
        "Prompt": prompt,
        "Response": response
    })

df = pd.DataFrame(results)
df.to_csv("ai_test_results.csv", index=False)

print("Results saved to ai_test_results.csv")
```
