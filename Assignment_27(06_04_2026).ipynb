### Image as Numbers
Load an image, print shape, pixel values, channels, and explain them.

```python
import cv2

# Load image
image = cv2.imread("sample.jpg")

# Print shape
print("Image Shape:", image.shape)

# Print pixel values (first pixel)
print("Pixel Value at (0,0):", image[0, 0])

# Print all pixel values (optional - large output)
# print(image)

# Number of channels
print("Number of Channels:", image.shape[2])
```
