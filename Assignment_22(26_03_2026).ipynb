### Movie Review Analyzer
Build a simple sentiment analyzer and test on 5 reviews.

```python
pip install TextBlob
```

```python
from textblob import TextBlob

reviews = [
    "Absolutely loved the cinematography and the music was शानदार!"
    "The movie started well but lost its charm halfway through."
    "Terrible script and poor direction ruined the entire experience."
    "A heartwarming story with brilliant performances by the cast."
    "Not worth the time, but the visuals were somewhat impressive."
]

for i, review in enumerate(reviews):
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity
    
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    print(f"Review {i+1}: {sentiment} (Polarity: {polarity:.2f})")
```
