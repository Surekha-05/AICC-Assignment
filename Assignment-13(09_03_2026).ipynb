### House Price Predictor 
##### Train a Linear Regression model, predict prices, and test with new input.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Size": [1000, 1500, 1800, 2400, 3000],
    "Bedrooms": [2, 3, 3, 4, 4],
    "Age": [10, 8, 5, 3, 2],
    "Price": [300000, 400000, 500000, 650000, 800000]

}

df = pd.DataFrame(data)
```

```python
X = df[["Size", "Bedrooms", "Age"]]
y = df["Price"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
new_house = [[2000, 3, 4]]  # Size, Bedrooms, Age
predicted_price = model.predict(new_house)
print("Predicted House Price:", predicted_price[0])
```
