## Keyword Extractor

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = [
    "Artificial intelligence is transforming the world",
    "Machine learning is a part of artificial intelligence",
    "Data science uses machine learning techniques"
]

# Create TF-IDF model
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Get feature names
features = vectorizer.get_feature_names_out()

# Display keywords for each document
for i, doc in enumerate(X):
    print(f"\nDocument {i+1} Keywords:")
    scores = zip(features, doc.toarray()[0])
    sorted_keywords = sorted(scores, key=lambda x: x[1], reverse=True)
    
    for word, score in sorted_keywords[:5]:
        print(word, round(score, 3))
```
