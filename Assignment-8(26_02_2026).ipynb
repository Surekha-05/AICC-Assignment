## Data Doctor
#### Clean a dataset by handling missing values, removing duplicates, standardizing text, and explain why cleaning matters.

```python
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv("data.csv")
```

```python
# Check missing values
print(data.isnull().sum())
```

```python
# Fill missing Salary with average salary
data["Salary"].fillna(data["Salary"].mean(), inplace=True)

# Fill missing Experience with median experience
data["Experience"].fillna(data["Experience"].median(), inplace=True)
```

```python
# Standardize text (make department lowercase)
data["Department"] = data["Department"].str.lower()
```

```python
# Remove duplicates (if any)
data = data.drop_duplicates()

# Show cleaned dataset
print(data)
```
