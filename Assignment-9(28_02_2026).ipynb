## Storytelling with Graphs
#### Description : Create bar chart, pie chart, histogram and write a short data story explaining trends..

```python
import matplotlib.pyplot as plt

employees = ['John','Sarah','David','Emma','Alex']
salary = [45000,52000,48000,60000,55000]

plt.bar(employees, salary)
plt.title("Employee Salary Comparison")
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.show()
```

```python
plt.pie(salary, labels=employees, autopct='%1.1f%%')
plt.title("Salary Distribution Among Employees")
plt.show()
```

```python
salary = [45000,52000,48000,60000,55000,47000,51000,59000,53000,46000]

plt.hist(salary, bins=5)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```

The bar chart shows the salary comparison among employees, where Emma has the highest salary (60000) while John has the lowest (45000).

The pie chart illustrates how the total salary is distributed among employees. Emma contributes the largest portion of the salary share, followed by Alex and Sarah.

The histogram shows that most employee salaries fall between 45,000 and 55,000, indicating that the majority of employees earn within a similar salary range.

Overall, the data suggests that salary distribution is fairly balanced, but a few employees like Emma earn higher due to experience or role.
