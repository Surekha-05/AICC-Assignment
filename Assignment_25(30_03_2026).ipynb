### Prompt Engineer
Write prompts for resume, business idea, study plan and compare weak vs strong prompts.

1. Resume Prompt
- Weak Prompt❌: “Create a resume for me.”
- Issue: The request is too general and lacks important details such as skills, job role, and preferred format.
- Strong Prompt✅ :
“Design a professional one-page resume for a final-year Computer Science student applying for a Data Analyst position. Include sections such as education, technical skills (Python, SQL, Excel), internships, and academic projects. Use strong action verbs, quantify achievements wherever possible, and ensure the format is clean and ATS-friendly.”

2. Business Idea Prompt
- Weak Prompt❌ : “Give me a business idea.”
- Issue: The prompt does not specify the industry, investment level, or target audience.
-Strong Prompt✅:
“Provide three innovative low-investment business ideas suitable for college students in India, focusing on technology or online-based services. For each idea, include the target audience, required resources, potential revenue model, and scalability opportunities.”

3. Study Plan Prompt
- Weak Prompt❌:“Create a study plan.”
- Issue: The prompt is unclear as it does not mention the subject, duration, or learning objectives.
- Strong Prompt✅:
“Develop a 4-week study plan for beginners to learn Python, assuming 2 hours of study per day. Include daily topics, hands-on exercises, and mini-projects. Emphasize fundamental concepts such as variables, loops, functions, and basic data analysis.”

| Aspect         | Weak Prompt                | Strong Prompt                                                    |
| -------------- | -------------------------- | ---------------------------------------------------------------- |
| Clarity        | Vague and unclear          | Specific and detailed                                            |
| Context        | No background info         | Includes role, goal, or situation                                |
| Output Quality | Generic results            | Tailored, high-quality output                                    |
| Instructions   | Minimal                    | Clear structure and expectations                                 |
| Example        | “Write something about AI” | “Explain AI in 300 words with real-world examples for beginners” |
