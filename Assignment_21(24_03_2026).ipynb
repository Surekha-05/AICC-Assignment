### Word Importance Explorer
Use TF-IDF on 5 documents and identify top keywords with explanation.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "AI is transforming industries with automation and intelligence",
    "Machine learning enables systems to learn from data",
    "Artificial intelligence and machine learning are closely related fields",
    "Data science uses machine learning for predictive analysis",
    "Automation improves efficiency in industries"
]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

for i, doc in enumerate(tfidf_matrix):
    print(f"\nDocument {i+1} Top Keywords:")
    scores = zip(feature_names, doc.toarray()[0])
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    for word, score in sorted_scores[:5]:
        print(f"{word}: {score:.3f}")
```

- Document 1: ai, transforming, industries, automation, intelligence
- Document 2: machine, learning, data, enables, systems
- Document 3: artificial, intelligence, machine, learning, related
- Document 4: data, science, predictive, analysis, learning
- Document 5: automation, efficiency, industries, improves

- Common words across documents (like “machine”, “learning”) get lower importance.
- Unique words (like “predictive”, “efficiency”) get higher importance.
- TF-IDF helps identify meaningful and context-specific keywords.
